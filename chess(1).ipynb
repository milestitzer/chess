```python
# Object-Oriented Chess by Miles Titzer

from IPython.display import clear_output
global boo,checkmate,opp_piece,piece2,castle
enpassant = False
castle = False
m=[]
pieces=[]
empty = '\033[1;31;48m   \033[1;30;49m'
before = False # establishes a boolean value that signals that before has not been established yet.
after = ''
piece = ''
bound = '  \033[1;30;49m|'+'---|'*8 #boundary design
bound2 = '  \033[1;30;49m+'+'---+'*8
colnames = 'abcdefgh'
rownames = [str(i) for i in range(1,9)]
boo = True
player = 'Blue'

# Temporary Matrix
t = ['empty' for row in range(0,8)]

# Pieces Matrix
for row in range(0,8):
    for col in range(0,8):
        piece = empty
        # I need to differentiate between each player's pieces, so I added \033 codes to alter fonts
        if row == 1:
            piece = '\033[1;31;48m P \033[1;30;49m'
        if row == 6:
            piece = '\033[1;34;48m P \033[1;30;49m'
        if row == 0:
            if col == 0 or col == 7:
                piece = '\033[1;31;48m R \033[1;30;49m'
            if col == 1 or col == 6:
                piece = '\033[1;31;48m N \033[1;30;49m'
            if col == 2:
                piece = "\033[1;31;48m B \033[1;30;49m"
            if col == 5:
                piece = "\033[1;31;48m B \033[1;30;49m"
        if row == 7:
            if col == 0 or col == 7:
                piece = '\033[1;34;48m R \033[1;30;49m'
            if col == 1 or col == 6:
                piece = '\033[1;34;48m N \033[1;30;49m'
            if col == 2:
                piece = "\033[1;34;48m B \033[1;30;49m"            
            if col == 5:
                piece = "\033[1;34;48m B \033[1;30;49m"            
        if row == 0 and col == 4:
            piece = '\033[1;31;48m K \033[1;30;49m'
        if row == 7 and col == 4:
            piece = '\033[1;34;48m K \033[1;30;49m'
        if row == 0 and col == 3:
            piece = '\033[1;31;48m Q \033[1;30;49m'
        if row == 7 and col == 3:
            piece = '\033[1;34;48m Q \033[1;30;49m'
               

        pieces.append(f'{piece}')
    t[row] = pieces
    pieces = []
pieces = t


# establish which coordinates contain white's pieces
welig = ['' for row in range(0,8)]
for row in range(0,8):
    m = []
    for col in range(0,8):
        if row < 6:
            m.append(False)
        else:
            m.append(True)
    welig[row] = m

belig = ['' for row in range(0,8)]
for row in range(0,8):
    m = []
    for col in range(0,8):
        if row > 1:
            m.append(False)
        else:
            m.append(True)
    belig[row] = m
        



# Board-Making Functions
def board(pieces,before,after):
    from IPython.display import clear_output
    clear_output
    
    
    
    # Board Matrix
    t = ['' for row in range(0,8)]
    if boo:
        print("BLUE'S TURN")
        print(bound2)
        for row in range(0,8):
            m = []
            for col in range(0,8):
                
                #make the pattern checkered
                pe = [char for char in pieces[row][col]] # list of piece elements
                pe[8] = '9' if abs(row-col)%2 == 0 else '7' # change background color
                pieces[row][col] = ''.join(pe)
                
                m.append(f'|{pieces[row][col]}')
            m = ''.join(m)
            t[row] = m
            print(f'\033[1;30;49m{8-row} \033[1;30;49m{m}|\033[1;30;49m  ')
            if row == 7:
                break
            print(bound)
        print(bound2)

        # Margin Labeling
        t = []
        for i in colnames:
            t.append(i)
        t = '   '.join(t)
        print(f'\033[1;30;49m    {t}  ')
    else:
        print("RED'S TURN")
        print(bound2)
        for row in range(0,8):
            m = []
            for col in range(0,8):
                
                #make the pattern checkered
                pe = [char for char in pieces[row][col]] # list of piece elements
                pe[8] = '9' if abs(row-col)%2 == 0 else '7' # change background color
                pieces[row][col] = ''.join(pe)

                m.append(f'|{pieces[7-row][7-col]}')
            m = ''.join(m)
            t[row] = m
            print(f'\033[1;30;49m{1+row} \033[1;30;49m{m}|\033[1;30;49m  ')
            if row == 7:
                break
            print(bound)
        print(bound2)

        # Margin Labeling
        t = []
        for i in colnames[::-1]:
            t.append(i)
        t = '   '.join(t)
        print(f'\033[1;30;49m    {t}  ')
        
    if not before: # if before has not been established yet, don't print nothin
        pass
    else:
        print(f'\n{player} moved their {piece[11]} from {before} to {after}')
        
    return pieces
        
# Dictionary for User Interface
idic = {
    'a':0,
    'b':1,
    'c':2,
    'd':3,
    'e':4,
    'f':5,
    'g':6,
    'h':7
}

# checks for wack coordinates 
def wack_beforecheck(before):
    global check1,check2,checkmate 
    checkmate = check_checkmate(before)
    if checkmate:
        return before,check1

    while len(before) != 2: # while input is invalid
        print("These coordinates are wack")
        check1 = False
        before = []
        before = input('move ')
        checkmate = check_checkmate(before)
        if checkmate: 
            return before,check1
    check1 = True
    if not check2:
        before,check2 = wack_beforecheck2(before)
    else:
        check1 = False
    return before,check1
        
        
def wack_beforecheck2(before):
    global check1,check2,checkmate

    while before[0] not in colnames or before[1] not in rownames: # while input is invalid
        print("These coordinates are wack")
        check2 = False
        before = []
        before = input('move ')
        checkmate = check_checkmate(before)
        if checkmate: 
            return before,check1
        
    check2 = True
    if not check1:
        before,check1 = wack_beforecheck(before)
    else:
        check2 = False

    return before,check2

def wack_aftercheck(before):
    global check1,check2,checkmate 
    checkmate = check_checkmate(before)
    checkmate
    if checkmate:
        return before,check1

    while len(before) != 2: # while input is invalid
        print("These coordinates are wack")
        
        check1 = False
        before = []
        before = input('to ')
        checkmate = check_checkmate(before)
        if checkmate: 
            return before,check1
    check1 = True
    if not check2:
        before,check2 = wack_beforecheck2(before)
    else:
        check1 = False
    return before,check1
        
        
def wack_aftercheck2(before):
    global check1,check2,checkmate

    while before[0] not in colnames or before[1] not in rownames: # while input is invalid
        print("These coordinates are wack")
        check2 = False
        before = []
        before = input('to ')
        checkmate = check_checkmate(before)
        if checkmate: 
            return before,check2
        
    check2 = True
    if not check1:
        before,check1 = wack_beforecheck(before)
    else:
        check2 = False

    return before,check2



rookmoved = [[False,False],[False,False]]
def rookmove(oppmove):
    if oppmove == (7,0):
        rookmoved[1][0] = True
    if oppmove == (7,7):
        rookmoved[1][1] = True
    if oppmove == (0,0):
        rookmoved[0][0] = True
    if oppmove == (0,7):
        rookmoved[0][1] = True
    return rookmoved 
    
global checkbool
checkbool = False
def check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig):

    checkbool = False
    
    # find the king
    if piece[11] == 'K':
        rowf_check = rowf
        colf_check = colf
    else:
        for rowf_check in range(0,8):
            for colf_check in range(0,8):
                if elig[rowf_check][colf_check] and pieces[rowf_check][colf_check][11] == 'K': break
            if elig[rowf_check][colf_check] and pieces[rowf_check][colf_check][11] == 'K': break

                          
    global checkloop
    checkloop = True  # stops moveking() from looping infinitely
    
    # if the king could be taken by one of its opponent's pieces
    # then the king is in check
    for r in range(0,8): 
        for c in range(0,8):             
            if nelig[r][c]:
                
                piece_check = pieces[r][c]
                leg1,piece = movepiece(r,c,rowf_check,colf_check,piece_check,rookmoved,not boo,belig,welig,nelig,elig,checkbool)
                
                elig[row][col] = False  # allows detection of uncovering a check
                elig[rowf][colf] = True                                        
                leg2 = obstructions(r,c,rowf_check,colf_check,piece_check,nelig,elig)
                
                if leg1 and leg2 and (r != rowf or c != colf): 
                # if the piece at (r,c) could attack king and is not being taken by pending move:
                    checkbool = True
                    print(f'Check from {pieces[r][c][11]}!')
                    break
                    
    checkloop = False
    elig[row][col] = True
    elig[rowf][colf] = False                                        
    
    
    return checkbool


def obstructions(row,col,rowf,colf,piece,elig,nelig):
    legal = True
    if piece[11] == 'K' or piece[11] == 'N':
        pass
    elif col == colf:
        if rowf > row:
            for r in range(row+1,rowf):
                if elig[r][col] or nelig[r][col]:
                    legal = False
        else:
            for r in range(rowf+1,row):
                if elig[r][col] or nelig[r][col] :
                    legal = False
    elif row == rowf:
        if colf > col:
            for c in range(col+1,colf):
                if elig[row][c] or nelig[row][c]:
                    legal = False
        else:
            for c in range(colf+1,col):
                if elig[row][c] or nelig[row][c]:
                    legal = False
    else: 
        slope = (rowf-row)/(colf-col)
        right = colf>col
        for x in range(1,abs(colf-col)):
            if right:
                if elig[int(row+x*slope)][int(col+x)] or nelig[int(row+x*slope)][int(col+x)]:
                    legal = False
            else:
                if elig[int(row-x*slope)][int(col-x)] or nelig[int(row-x*slope)][int(col-x)]:
                    legal = False
    return legal


# King's Legality Matrix
global wkingmoved,bkingmoved
wkingmoved = False
bkingmoved = False
bot = 0
def moveking(row,col,rowf,colf,rookmoved,elig,nelig,boo,wkingmoved,bkingmoved,piece):
    
    global rookmove,castle,piece2,rookrow,rookcol,rookcolf
    
    legal = False
    checkbool = False
    if abs(rowf-row)<=1 and abs(colf-col)<=1:
        legal = True
    
    # if checkloop: print('checking!')

    elif abs(colf-col)==2 and not checkloop:
        castle = False
        #pdb.set_trace()
        
        if boo:
            if not wkingmoved:
                if colf>col:
                    if not rookmoved[1][1]:
                        if not any(elig[row][col+1:6]):
                            if not any(nelig[row][col:6]):
                                savecolf = colf
                                for colf in range(col,7):
                                    checkbool = check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig)
                                    if checkbool:
                                        print('Check!')
                                        checkbool = False
                                        
                                if not checkbool:
                                    legal = True
                                    wkingmoved = True
                                    castle = True
                                    rookrow = 7; rookcol = 7; rookcolf = 5
                            colf = savecolf

                else:
                    if not rookmoved[1][0]:
                        if not any(elig[row][1:col]):
                            if not any(nelig[row][1:col]):
                                savecolf = colf
                                for colf in range(1,col+1):

                                    checkbool = check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig)
                                    if checkbool:
                                        print('Check!')
                                        checkbool = False
                                        
                                if not checkbool:
                                    legal = True
                                    bkingmoved = True
                                    castle = True
                                    rookrow = 7; rookcol = 0; rookcolf = 3
                                colf = savecolf
        else:
            if not bkingmoved:
                if colf>col:
                    if not rookmoved[0][1]:
                        if not any(elig[row][col+1:6]):
                            if not any(nelig[row][col+1:6]):
                                savecolf = colf
                                for colf in range(col,7):
                                    checkbool = check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig)
                                    if checkbool:
                                        print('Check!')
                                        checkbool = False
                                        
                                if not checkbool:
                                    legal = True
                                    bkingmoved = True      
                                    castle = True
                                    rookrow = 0; rookcol = 7; rookcolf = 5
                                colf = savecolf  
                else:
                    if not rookmoved[0][0]:
                        if not any(elig[row][1:col]):
                            if not any(nelig[row][1:col]):
                                savecolf = colf
                                for colf in range(1,col+1):
                                    checkbool = check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig)
                                    if checkbool:
                                        print('Check!')
                                        checkbool = False
                                        
                                if not checkbool:
                                    legal = True
                                    bkingmoved = True      
                                    castle = True
                                    rookrow = 0; rookcol = 0; rookcolf = 5
                                colf = savecolf    
        if castle: 
            print('Castle!')
            piece2 = pieces[rookrow][rookcol]
    return legal,piece
                      
# Queen's Legality Matrix
def movequeen(row,col,rowf,colf,piece,elig,nelig):
    legal = False
    if abs(col-colf) == abs(row - rowf):
        legal = True
    elif abs(col-colf) == 0:
        legal = True
    elif 0 == abs(row - rowf):
        legal = True
    return legal,piece

# Rook's Legality Matrix
def moverook(row,col,rowf,colf,piece,elig,nelig):
    legal = False
    if abs(col-colf) == 0:
        legal = True
    elif 0 == abs(row - rowf):
        legal = True
    return legal,piece
        
# Bishop's Legality Matrix
def movebishop(row,col,rowf,colf,piece,elig,nelig):

    legal = False
    if abs(col-colf) == abs(row - rowf):
        legal = True
    return legal,piece

# Knight's Legality Matrix
def moveknight(row,col,rowf,colf,piece,elig,nelig):
    legal = False
    if (abs(col-colf) == 1 and abs(row - rowf) == 2) or (abs(col-colf) == 2 and abs(row - rowf) == 1):
        legal = True
    return legal,piece
        
# Pawn's Legality Matrix
def movepawn(row,col,rowf,colf,boo,elig,nelig,piece,oppmove):
    global enpassant
    plist = ['K','Q','P','B','N','R']
    legal = False


    if boo: one = -1
    else: one = 1
    if col == colf and rowf-row == one and not nelig[rowf][colf]: 
        legal = True      
    elif col == colf and (row == 1 or row == 6) and abs(rowf - row) == 2:
        legal = True
    elif nelig[rowf][colf] and abs(colf-col) == 1 and rowf - row == one:
        legal = True
    elif oppmove==(row,colf) and row==3.5+one*.5 and opp_piece[11]=='P' and abs(colf-col)==1 and rowf-row==one:
        legal = True
        enpassant = True
        print('En Passant!')
    if legal and (rowf == 0 or rowf == 7):
        pdb.set_trace()
        p = ' '
        while p not in plist:
            p = input('Swap pawn for ')   
        piece_temp = piece.split()
        piece_temp[1] = ' '+p+' '
        piece_temp = ''.join(piece_temp)
        piece = piece_temp

    return legal,piece
        
def movepiece(row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig,checkbool):
    Ldic = {
        'K': moveking(row,col,rowf,colf,rookmoved,elig,nelig,boo,wkingmoved,bkingmoved,piece),
        'Q': movequeen(row,col,rowf,colf,piece,elig,nelig),
        'B': movebishop(row,col,rowf,colf,piece,elig,nelig),
        'R': moverook(row,col,rowf,colf,piece,elig,nelig),
        'N': moveknight(row,col,rowf,colf,piece,elig,nelig),
        'P': movepawn(row,col,rowf,colf,boo,elig,nelig,piece,oppmove)  
            }
    legal,piece = Ldic[piece[11]]
    return legal,piece
        
# Move Legality
def legality(pieces,elig,nelig,idic,checkbool):

    after=''; row=''; col=''; rowf=''; colf=''; piece=''
    
    global checkmate
    legal = False   

    while not legal:
        before = []
        before = input('move ')
        checkmate = check_checkmate(before)
        if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

        before,check1 = wack_beforecheck(before)
        checkmate = check_checkmate(before)
        if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

        row = 8-int(before[1])
        col = idic[before[0]]

        
        while not elig[row][col]:
            print("These coordinates don't have an eligible piece")
            before = []
            before = input('move ')
            checkmate = check_checkmate(before)
            if checkmate: break

            before,check1 = wack_beforecheck(before)
            checkmate = check_checkmate(before)
            if checkmate: break

            row = 8-int(before[1])
            col = idic[before[0]]
            
            
        after = [] 
        after = input('to ')
        checkmate = check_checkmate(after)
        if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

        after,check1 = wack_aftercheck(after)
        checkmate = check_checkmate(before)
        if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

        rowf = 8-int(after[1])
        colf = idic[after[0]]

        
        while elig[rowf][colf]:
            print("Silly user, you can't take your own piece!")
            before = []
            before = input('move ')
            checkmate = check_checkmate(before)
            if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

            before,check1 = wack_beforecheck(before)
            checkmate = check_checkmate(before)
            if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

            row = 8-int(before[1]) 
            col = idic[before[0]]

            after = []
            after = input('to ')
            checkmate = check_checkmate(after)
            if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

            after,check1 = wack_aftercheck(after)
            checkmate = check_checkmate(before)
            if checkmate: return legal,before,after,checkbool,row,col,rowf,colf,piece

            # rewrite coordinates
            rowf = 8-int(after[1])
            colf = idic[after[0]]        
        
        
        piece = pieces[row][col]
        leg1,piece = movepiece(row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig,checkbool)
        leg2 = obstructions(row,col,rowf,colf,piece,elig,nelig)        
        
        
        if leg1 and leg2: # if move is within range of piece and unobstructed
            
            checkbool = check(pieces,legal,row,col,rowf,colf,piece,rookmoved,boo,belig,welig,elig,nelig) 
            
            if not checkbool:
                legal = True
                break
            else: print("Oops, looks like that move would put you in check.")
                
        else: print("Sorry, but that wasn't a legal move.")
                            
    return legal,before,after,checkbool,row,col,rowf,colf,piece

def check_checkmate(before):
    global checkmate
    if before == 'resign':
        checkmate = True
        return checkmate
        


        
############# Gameplay ################


checkmate = False
check1 = False
check2 = False
oppmove = (3,3)
opp_piece = ' '

#plist,plistvalues,pdic = pieces_list()
print('Enter the coordinates (e.g. e2) of the piece you would like to move.')
while not checkmate:
    
    pieces = board(pieces,before,after) # print the board
    # call the interface dictionary
    rookmove(oppmove) # update rookmoved to check castling possibility

    print('\033[0;30;49m\nInput the coordinates of the piece you would like to move.')
    print('To resign, just type resign.')

    if boo:          # if it's white's turn
        elig = welig
        nelig = belig
    else:
        elig = belig
        nelig = welig

    legal,before,after,checkbool,row,col,rowf,colf,piece = legality(pieces,elig,nelig,idic,checkbool)
    if checkmate: break
        
    # Update pieces, elig, oppmove
    pieces[rowf][colf] = piece # put the piece in its new spot
    elig[rowf][colf] = True
    nelig[rowf][colf] = False
    oppmove = (rowf,colf)
    opp_piece = piece
    
    # put a blank space where the piece was:
    blankspace = piece.split()
    blankspace[1] = '   '
    blankspace = ''.join(blankspace)
    pieces[row][col] = blankspace
    elig[row][col] = False
    
    if enpassant: 
        nelig[row][colf] = False
        blankspace = piece.split()
        blankspace[1] = '   '
        blankspace = ''.join(blankspace)
        pieces[row][colf] = blankspace

    if castle:
        row = rookrow; rowf = rookrow; col = rookcol; colf = rookcolf
        pieces[rowf][colf] = piece2
        elig[rowf][colf] = True
        nelig[rowf][colf] = False
        
        blankspace = piece.split()
        blankspace[1] = '   '
        blankspace = ''.join(blankspace)
        pieces[row][col] = blankspace
        elig[row][col] = False
    
        

    if boo:
        welig = elig
        belig = nelig
    else:
        belig = elig
        welig = nelig
        
    if boo:
        player = 'Blue'
    else:
        player = 'Red'
        
    boo = not boo
    legal = False 
    enpassant = False
    castle = False
    rookrow = ''; rookcol = ''; rookcolf = ''
        
# Upon termination of while loop:
boo = not boo
print(f'{player} wins!') 
```
